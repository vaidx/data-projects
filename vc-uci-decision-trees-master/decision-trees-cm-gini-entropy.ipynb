```python
import pandas as pd
import numpy as np
import datetime
from sklearn import linear_model
from sklearn import tree
from sklearn import model_selection
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import graphviz
from graphviz import Source
from sklearn.metrics import confusion_matrix
from confusion_matrix_pretty_print import *
import warnings
warnings.filterwarnings('ignore')
```

```python
table = pd.read_csv('VertebralColumnDataset.csv')
```

```python
table.class_label.value_counts()
```

```python
X= pd.DataFrame()
```

```python
X['pelvic_incidence'] = table['pelvic_incidence']
X['pelvic_tilt'] = table['pelvic_tilt']
X['lumbar_lordosis_angle'] = table['lumbar_lordosis_angle']
X['sacral_slope'] = table['sacral_slope']
X['pelvic_radius'] = table['pelvic_radius']
X['degree_spondylolisthesis'] = table['degree_spondylolisthesis']
X['class_label'] = table['class_label']
```

```python
LabelE = LabelEncoder()
y = LabelE.fit_transform(X['class_label'])
```

```python
X = X.drop(['class_label'], axis = 1)
print(list(X.columns.values))
print(y)

feature_name_list = open('VertebralColumnDataset.csv').readlines()[0][:-2].split(',')[0:6]

X_train1, X_test1, y_train1, y_test1= train_test_split(X, y, test_size = 0.2, random_state = 1234,stratify=y)
X_train2, X_test2, y_train2, y_test2= train_test_split(X, y, test_size = 0.3, random_state = 1234, stratify=y)
X_train3, X_test3, y_train3, y_test3= train_test_split(X, y, test_size = 0.4, random_state = 1234,stratify=y)
```

```python
#for tree 1 with 20% Testing - ENTROPY
dtree_model02 = DecisionTreeClassifier(criterion='entropy',max_depth=10,max_leaf_nodes=7)
dtree_model02.fit(X_train1, y_train1)
dtree_predictions02 = dtree_model02.predict(X_test1)
score02 = accuracy_score(y_test1, dtree_predictions02)
print("Accuracy is "+str(score02*100))


#for tree 2 with 30% Testing - ENTROPY
dtree_model03 = DecisionTreeClassifier(criterion='entropy',max_depth=10, max_leaf_nodes=7)
dtree_model03.fit(X_train2, y_train2)
dtree_predictions03 = dtree_model03.predict(X_test2)
score03 = accuracy_score(y_test2, dtree_predictions03)
print("Accuracy is "+str(score03*100))

#for tree 3 with 40% Testing -ENTROPY
dtree_model04 = DecisionTreeClassifier(criterion='entropy',max_depth=10,max_leaf_nodes=7)
dtree_model04.fit(X_train3, y_train3)
dtree_predictions04 = dtree_model04.predict(X_test3)
score04 = accuracy_score(y_test3, dtree_predictions04)
print("Accuracy is "+str(score04*100))

#for tree 1 with 20% Testing - GINI
dtree_model12 = DecisionTreeClassifier(criterion='gini',max_depth=10,max_leaf_nodes=7)
dtree_model12.fit(X_train1, y_train1)
dtree_predictions12 = dtree_model12.predict(X_test1)
score12 = accuracy_score(y_test1, dtree_predictions12)
print("Accuracy is "+str(score12*100))


#for tree 2 with 30% Testing - GINI
dtree_model13 = DecisionTreeClassifier(criterion='gini',max_depth=10, max_leaf_nodes=7)
dtree_model13.fit(X_train2, y_train2)
dtree_predictions13 = dtree_model13.predict(X_test2)
score13 = accuracy_score(y_test2, dtree_predictions13)
print("Accuracy is "+str(score13*100))

#for tree 3 with 40% Testing -GINI
dtree_model14 = DecisionTreeClassifier(criterion='gini',max_depth=10,max_leaf_nodes=7)
dtree_model14.fit(X_train3, y_train3)
dtree_predictions14 = dtree_model14.predict(X_test3)
score14 = accuracy_score(y_test3, dtree_predictions14)
print("Accuracy is "+str(score14*100))
```

```python
for y in range(5,6):
    for x in range(4,11):
        #for tree 1 with 20% Testing - ENTROPY
        dtree_model02 = DecisionTreeClassifier(criterion='entropy',max_depth=x)
        dtree_model02.fit(X_train1, y_train1)
        dtree_predictions02 = dtree_model02.predict(X_test1)
        score02 = accuracy_score(y_test1, dtree_predictions02)
        print("Accuracy is "+str(score02*100)+"   x= "+str(x)+ " y= "+str(y)+" 20% ENTROPY")
        print(" ")


        #for tree 2 with 30% Testing - ENTROPY
        dtree_model03 = DecisionTreeClassifier(criterion='entropy',max_depth=x)
        dtree_model03.fit(X_train2, y_train2)
        dtree_predictions03 = dtree_model03.predict(X_test2)
        score03 = accuracy_score(y_test2, dtree_predictions03)
        print("Accuracy is "+str(score03*100)+"   x= "+str(x)+ " y= "+str(y)+" 30% ENTROPY")
        print(" ")

        #for tree 3 with 40% Testing -ENTROPY
        dtree_model04 = DecisionTreeClassifier(criterion='entropy',max_depth=x)
        dtree_model04.fit(X_train3, y_train3)
        dtree_predictions04 = dtree_model04.predict(X_test3)
        score04 = accuracy_score(y_test3, dtree_predictions04)
        print("Accuracy is "+str(score04*100)+"   x= "+str(x)+ " y= "+str(y)+" 40% ENTROPY")
        print(" ")
        
        #for tree 1 with 20% Testing - GINI
        dtree_model12 = DecisionTreeClassifier(criterion='gini',max_depth=x)
        dtree_model12.fit(X_train1, y_train1)
        dtree_predictions12 = dtree_model12.predict(X_test1)
        score12 = accuracy_score(y_test1, dtree_predictions12)
        print("Accuracy is "+str(score12*100)+"   x= "+str(x)+ " y= "+str(y)+" 20% GINI")
        print(" ")

        #for tree 2 with 30% Testing - GINI
        dtree_model13 = DecisionTreeClassifier(criterion='gini',max_depth=x)
        dtree_model13.fit(X_train2, y_train2)
        dtree_predictions13 = dtree_model13.predict(X_test2)
        score13 = accuracy_score(y_test2, dtree_predictions13)
        print("Accuracy is "+str(score13*100)+"   x= "+str(x)+ " y= "+str(y)+" 30% GINI")
        print(" ")

        #for tree 3 with 40% Testing -GINI
        dtree_model14 = DecisionTreeClassifier(criterion='gini',max_depth=x)
        dtree_model14.fit(X_train3, y_train3)
        dtree_predictions14 = dtree_model14.predict(X_test3)
        score14 = accuracy_score(y_test3, dtree_predictions14)
        print("Accuracy is "+str(score14*100)+"   x= "+str(x)+ " y= "+str(y)+" 40% GINI")
        print(" ")
```
